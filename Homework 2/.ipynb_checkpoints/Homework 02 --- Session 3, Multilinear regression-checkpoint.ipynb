**Answerï¼š**    
Since the dummy variable usually represents a mutex, a value of 0 causes the coefficient of the variable to have no effect on the dependent variable, and when the dummy value is 1, its coefficient is used to change the intercept. Therefore, from the table, x3 is serves as dummy variable.

* (b) Calculate the coefficients $\hat w$ manually or with python *numpy* module by using matrix computation.

```python
X = np.matrix(ex1[['x3','x1','x2']])
Y = np.matrix(ex1.y).T
w = (X.T * X).I * X.T * Y
print(w)
```

* (c) Verify your answer by implementing the OLS regression function with python *statsmodels* module.

```python
lm1_ex1 = smf.ols(formula = 'y~x1+x2',data=ex1).fit()
print(lm1_ex1.params)
```

```python
w1 = []
[w1.append(x[0]) for x in w.tolist()]
pd.DataFrame({'Matrix_Computation':w1,'OLS Regression':lm1_ex1.params})
```

* (d) Regress over these proportions (the ones mentioned in part b, leave out "No Insurance Coverage") to predict income -- What interesting results do we have here? Briefly state what you've obtained.

#### Extra credit:
* (e) Visualize the observed income versus the predicted income, like what we did in the lab session. Does this fitted line look great or not? If you're not sure, check the correlation matrix and make a guess.

```python
ex2['TotalIncome'] = ex2.IncomePerCapita * ex2.TotalPop
```

```python
train, valid = train_test_split(ex2, test_size = 0.3)
```

```python
lm1_ex2 = smf.ols(formula = 'TotalIncome~TotalPop+PrivateOnly+PublicOnly+BothPublicAndPrivate',data=train).fit()
```

```python
def modelEval(lm, key = 'TotalIncome'):
    lmy = lm.predict(valid)
    y_err = lmy - valid[key]
    y_norm = valid[key]-mean(valid[key])
    R2 = 1-y_err.dot(y_err)/y_norm.dot(y_norm)  # R2 w.r.t. validation set
    print('Validation R^2 is %f'%R2)
    print(lm.summary())
    plt.plot(lmy, lmy, 'b-')
    plt.plot(lmy, valid[key],'or')
```

```python
modelEval(lm1_ex2)
```

**Since the p-value of PublicOnly is much higher than others,I would try to remove it.**

```python
lm2_ex2 = smf.ols(formula = 'TotalIncome~TotalPop+PrivateOnly+BothPublicAndPrivate',data=ex2).fit()
```

```python
modelEval(lm2_ex2)
```

Now p-value is tiny enough(p < 5% ) that we can take a conclusion:

The impact of the Total Population, Private Only and both Private&Public insurance population are statistically significant (p-value < 5%) on income.

In addition, the validation R2 is improved so the fit is good

* (e) Create a figure with four (4) subfigures, one for each polynomial degree M = 1, 2, 3, 4. In each subfigure:

```python
fig = pl.figure(figsize=[12,12])
g = lambda x: (9*(x**2) - x +1)
for M in [1,2,3,4]:
    ax = fig.add_subplot(2,2,M)
    ax.scatter(ex3.x1,ex3.y,label='original x1 & y',c='r')
    polyfit = np.polyfit(ex3.x1, ex3.y, M)
    f = np.poly1d(polyfit)
    x_new = np.linspace(ex3.x1.min(), ex3.x1.max(), 10000)
    y_new = f(x_new)
    lmy = f(ex3.x1)
    y_err = lmy - ex3.y
    y_norm = ex3.y - np.mean(ex3.y)
    R2 = 1-y_err.dot(y_err)/y_norm.dot(y_norm)
    ax.plot(x_new,g(x_new),label='true function',c='b')
    ax.plot(x_new,y_new,label='polynomial fit',c='g')    
    ax.legend()
    formula = 'y = '+ str(round(f.coeffs[-1],4)) 
    for i in range(len(f),0,-1):
        formula = formula + ' + '+ str(round(f.coeffs[len(f)-i],4)) + '$x^'+str(i)+'$'
    ax.set_title('M = {}\n $R^2$ = {} \n {}'.format(M,R2,formula))
fig.suptitle('Fitting for y and x1', y = 0.92,fontsize=16)
```

```python
fig = pl.figure(figsize=[16,16])
g = lambda x: (9*(x**2) - x +1)
for M in [1,2,3,4]:
    ax = fig.add_subplot(2,2,M)
    ax.scatter(ex3.x1,ex3.y,label='original x1 & y',c='r')
    polyfit = np.polyfit(ex3.x1, ex3.y, M)
    f = np.poly1d(polyfit)
    np.random.seed(999)
    x0 = np.random.uniform(np.min(ex3.x1),np.max(ex3.x1))
    x_new = np.linspace(ex3.x1.min(), ex3.x1.max(), 10000)
    y_new = f(x_new)
    lmy = f(ex3.x1)
    y_err = lmy - ex3.y
    y_norm = ex3.y - np.mean(ex3.y)
    R2 = 1-y_err.dot(y_err)/y_norm.dot(y_norm)
    ax.plot(x0,g(x0),'yd',markersize=12,label='x picked on true function')
    ax.plot(x0,f(x0),'kd',markersize=12,label='x picked on fitted function')
    ax.plot([x0,x0],[g(x0),f(x0)],'--',label='prediction error')
    prediction_error = g(x0)-f(x0)
    ax.plot(x_new,g(x_new),label='true function',c='b')
    ax.plot(x_new,y_new,label='polynomial fit',c='g')    
    ax.legend()
    formula = 'y = '+ str(round(f.coeffs[-1],4)) 
    for i in range(len(f),0,-1):
        formula = formula + ' + '+ str(round(f.coeffs[len(f)-i],4)) + '$x^'+str(i)+'$'
    ax.set_title('M = {}\n $R^2$ = {} \n {}\n Prediction Error = {}'.format(M,R2,formula,prediction_error))
fig.suptitle('Fitting for y and x1', y = 0.92,fontsize=16)
```

    * Compute the average sum of squares of errors $E = \frac{\sum_i^n (y_i - f(x_i))^2}{n}$ ($x_i$ denotes your future points), mention it in your subfigure's title (**Be sure to use the SAME future points for ALL 4 models when you calculate prediction error**)

```python
fig = pl.figure(figsize=[20,20])
g = lambda x: (9*(x**2) - x +1)
for M in [1,2,3,4]:
    ax = fig.add_subplot(2,2,M)
    ax.scatter(ex3.x1,ex3.y,label='original x1 & y',c='r')
    polyfit = np.polyfit(ex3.x1, ex3.y, M)
    f = np.poly1d(polyfit)
    np.random.seed(999)
    prediction_error = []
    yi = []
    xi = []
    fxi = []
    for i in range(10):
        x0 = np.random.uniform(np.min(ex3.x1),np.max(ex3.x1))
        ax.plot(x0,g(x0),'yd',markersize=12,label='x picked on true function')
        ax.plot(x0,f(x0),'kd',markersize=12,label='x picked on fitted function')
        ax.plot([x0,x0],[g(x0),f(x0)],'--',label='prediction error')
        prediction_error.append(-g(x0)+f(x0))
        xi.append(x0)
        yi.append(g(x0))
        fxi.append(f(x0))
    x_new = np.linspace(ex3.x1.min(), ex3.x1.max(), 10000)
    y_new = f(x_new)
    lmy = f(ex3.x1)
    y_err = lmy - ex3.y
    y_norm = ex3.y - np.mean(ex3.y)
    R2 = 1-y_err.dot(y_err)/y_norm.dot(y_norm)
    prediction_error = np.mean(prediction_error)
    E = np.sum((np.array(fxi) - np.array(yi))**2)/10
    ax.plot(x_new,g(x_new),label='true function',c='b')
    ax.plot(x_new,y_new,label='polynomial fit',c='g')    
    # Make label only one
    handles, labels = pl.gca().get_legend_handles_labels()
    by_label = OrderedDict(zip(labels, handles))
    pl.legend(by_label.values(), by_label.keys())
    formula = 'y = '+ str(round(f.coeffs[-1],4)) 
    
    for i in range(len(f),0,-1):
        formula = formula + ' + '+ str(round(f.coeffs[len(f)-i],4)) + '$x^'+str(i)+'$'
    ax.set_title('M = {}\n $R^2$ = {} \n {}\n Mean Prediction Error = {}\n E = {}'.format(M,R2,formula,prediction_error,E))
fig.suptitle('Fitting for y and x1', y = 0.92,fontsize=16)
```
